# Exploration des données

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données depuis votre dataframe (assurez-vous que votre dataframe s'appelle 'df')
# df = pd.read_csv('nom_du_fichier.csv')  # Assurez-vous de spécifier le bon chemin si votre fichier est dans un autre répertoire

# Afficher les premières lignes du jeu de données pour avoir un aperçu
# print(df.head())

# Visualiser la distribution des variables numériques
numerical_features = ['1stFlrSF', '2ndFlrSF', 'BedroomAbvGr', 'FullBath']
df[numerical_features].hist(bins=20, figsize=(10,8))
plt.suptitle('Distribution des variables numériques', x=0.5, y=1.02, ha='center', fontsize='large')
plt.show()

# Visualiser les relations entre les variables continues avec des scatter plots
sns.pairplot(df[numerical_features])
plt.suptitle('Relations entre les variables continues', x=0.5, y=1.02, ha='center', fontsize='large')
plt.show()

# Visualiser les distributions des variables catégoriques avec des graphiques à barres
categorical_features = ['MSZoning', 'Street', 'Alley', 'Utilities']  # Choisissez les variables catégoriques pertinentes
for feature in categorical_features:
    sns.countplot(x=feature, data=df)
    plt.title(f'Distribution de {feature}')
    plt.xticks(rotation=45)
    plt.show()

# Créer une matrice de corrélation pour identifier les relations linéaires entre les variables numériques
correlation_matrix = df[numerical_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matrice de corrélation')
plt.show()

# Identifier les valeurs manquantes à l'aide d'une carte de chaleur
sns.heatmap(df.isnull(), cbar=False)
plt.title('Carte de chaleur des valeurs manquantes')
plt.show()
```

# Prétraitement des données 

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Charger les données depuis votre dataframe (assurez-vous que votre dataframe s'appelle 'df')
# df = pd.read_csv('nom_du_fichier.csv')  # Assurez-vous de spécifier le bon chemin si votre fichier est dans un autre répertoire

# Traitement des valeurs manquantes
imputer = SimpleImputer(strategy='mean')  # Remplace les valeurs manquantes par la moyenne de chaque colonne
df_imputed = pd.DataFrame(imputer.fit_transform(df[numerical_features]), columns=numerical_features)  # Applique l'imputation aux colonnes numériques
df[categorical_features] = df[categorical_features].fillna('Unknown')  # Remplace les valeurs manquantes dans les colonnes catégoriques par 'Unknown'

# Normalisation des caractéristiques numériques
scaler = StandardScaler()  # Initialise le scaler pour la normalisation
df_scaled = pd.DataFrame(scaler.fit_transform(df_imputed), columns=numerical_features)  # Applique la normalisation aux données numériques

# Concaténation des données traitées
df_processed = pd.concat([df_scaled, df[categorical_features]], axis=1)  # Concatène les données numériques normalisées avec les données catégoriques non modifiées

# Affichage des premières lignes du jeu de données traité pour vérification
print(df_processed.head())

# Vérification des données après prétraitement
print(df_processed.info())
```
